# 🍽️ MiamLocal — Du compte-rendu aux user stories

MiamLocal est une application de livraison de repas en circuit court, fondée à Lyon en 2022. Contrairement aux grandes plateformes généralistes, MiamLocal se positionne sur les restaurateurs indépendants et les producteurs locaux : chaque restaurant partenaire est situé à moins de 15 kilomètres du client, les livreurs sont salariés (pas auto-entrepreneurs), et la commission prélevée aux restaurateurs est plafonnée à 15% contre 25 à 30% chez les concurrents. L'application compte aujourd'hui 45 000 utilisateurs actifs et 320 restaurants partenaires répartis sur Lyon, Grenoble et Saint-Étienne, avec une levée de fonds en Série A de 4 millions d'euros bouclée en mars 2024. Le pari de MiamLocal, c'est que la proximité et l'équité du modèle suffisent à fidéliser à la fois les restaurateurs et les clients. Ce pari est aujourd'hui fragilisé par un taux de churn restaurateurs en hausse et une concurrence qui commence à copier le discours "local".

## Exercice 1 — Extraction structurée


### Contexte

Le PO sort de réunion. Il a son CR brut. Première chose à faire : en extraire les éléments actionnables. C'est le travail le plus mécanique — et celui où le LLM est immédiatement utile.

### Consigne

Copiez le prompt ci-dessous dans votre LLM, **suivi du CR complet**.

### ✏️ Prompt à copier-coller

> ```
> Tu es un Product Owner expérimenté. Analyse le compte-rendu
> de réunion ci-dessous et extrais trois listes :
>
> 1. DÉCISIONS ACTÉES : uniquement ce qui a été explicitement
>    validé par les participants. Si une « décision » est
>    ambiguë ou conditionnelle, classe-la en point ouvert.
>
> 2. ACTIONS : avec [Responsable] [Délai si mentionné]
>    [Dépendance éventuelle]
>
> 3. POINTS OUVERTS : questions non tranchées, désaccords
>    non résolus, informations manquantes.
>
> Sois rigoureux : ne classe en « décision » que ce qui est
> sans ambiguïté.
>


### 🔍 Grille d'analyse — Vérifiez dans la réponse du LLM

| Point de contrôle | Appréciation |
|---|---|
| Le LLM a-t-il classé « on tranchera » (mécanisme de récompense) en **point ouvert** et non en décision ? | |
| A-t-il détecté que la deadline « septembre » est posée par Marie mais **non validée** par le CTO ? | |
| Le ticket `#ML-2089` (dashboard temps réel) est-il identifié comme **déjà existant** dans le backlog ? | |
| L'absence de Alexis apparaît-elle comme une **dépendance bloquante** ? | |
| Les actions ont-elles un **responsable nommé** (pas juste « l'équipe ») ? | |

## Exercice 2 — Détection des risques et contradictions


### Contexte

Un bon PO lit un CR pour ce qu'il dit, mais aussi ce qu'il ne dit pas. Les risques sont souvent implicites, entre les lignes, dans les silences, dans les changements de sujet. Le LLM peut aider à les révéler.

### Consigne

Copiez le prompt ci-dessous dans votre LLM, suivi du CR. Notez le mot-clé **« auditeur externe »** — il change la posture du modèle.

### ✏️ Prompt à copier-coller

> ```
> Analyse ce compte-rendu de comité produit en tant qu'auditeur
> externe. Identifie :
>
> - Les CONTRADICTIONS entre ce qui est dit par différents
>   intervenants
> - Les RISQUES IMPLICITES que personne ne semble avoir relevés
> - Les DÉCALAGES entre les attentes (délais, périmètre,
>   ressources)
> - Les DÉPENDANCES NON TRAITÉES
>
> Pour chaque élément, cite le passage du CR qui le révèle
> et formule le risque de manière explicite.
>


### 🔍 Réponses attendues (exemples)

| Risque | Détail |
|--------|--------|
| **Ticket fermé / feature cassée** | Notifs push marquées « fermé » mais ne fonctionnent pas sur iOS. Annoncé aux restaurateurs → perte de confiance. |
| **Deadline non validée** | Marie impose septembre. Julien chiffre la fidélité à « un trimestre ». Amélie a besoin des maquettes 3 semaines avant. L'équation ne tient pas. |
| **Personne clé absente** | Alexis en congés sans remplaçant. Bloque la branche iOS, la migration Stripe, le pipeline temps réel. |
| **Estimation élastique** | Julien dit « 2-3 jours » pour le fix iOS, puis « c'est pas un truc qu'on case » pour Stripe (5 jours). Double standard. |
| **Conflit de ressources** | Théo affecté à la fidélité ET candidat pour Stripe. UX researcher à mi-temps et déjà occupée. |
| **Dette technique reportée** | Stripe déprécié en octobre, reporté au sprint 25. « C'était loin en janvier aussi » (Julien). |

## Exercice 4 — Questions pour le sprint planning


### Contexte

Simon prépare le sprint planning du sprint 24. Avant de s'engager devant l'équipe, il doit lever les zones d'ombre. Le LLM peut l'aider à formuler les bonnes questions aux bonnes personnes.

> ```

## Exercice 6 — Vérification d'une synthèse

### Contexte

Utiliser le LLM non pas comme **générateur** mais comme **vérificateur**.

En situation réelle, un PO reçoit souvent des synthèses écrites par d'autres (chef de projet, manager, consultant). Il doit détecter les déformations, omissions et surinterprétations.

### La "mauvaise" synthèse

Voici une synthèse rédigée par un chef de projet fictif. Elle contient des erreurs, des déformations et des omissions. Lisez-la attentivement.

> **Synthèse Comité Produit #14** — *rédigée par Paul Garnier, Chef de projet*
>
> Le comité produit du 12 juin a permis de valider plusieurs avancées importantes.
>
> Le sprint 23 est un succès complet : la refonte du tunnel de commande a doublé le taux de conversion mobile, et les notifications push géolocalisées sont désormais opérationnelles sur l'ensemble du parc.
>
> Le programme de fidélité a été validé à l'unanimité. Il permettra aux restaurateurs de créer des campagnes promotionnelles personnalisées avec un moteur de règles avancé. Le lancement est prévu pour septembre. L'équipe technique confirme la faisabilité du calendrier.
>
> Par ailleurs, une fonctionnalité innovante de recherche par « mood » sera intégrée dans la prochaine version, utilisant l'intelligence artificielle pour recommander des plats en fonction de l'humeur de l'utilisateur.
>
> Le sprint 24 se concentrera sur le programme de fidélité. La migration du système de paiement est terminée.
>
> Aucun risque majeur n'a été identifié.

### Consigne

Copiez le prompt ci-dessous dans votre LLM. Vous devrez y coller **les deux documents** : le CR original et la mauvaise synthèse ci-dessus.

### ✏️ Prompt à copier-coller

> ```
> Je te fournis deux documents :
> - DOCUMENT A : le compte-rendu détaillé d'un comité produit
> - DOCUMENT B : une synthèse rédigée par un chef de projet
>
> Compare B à A et identifie :
>
> 1. ERREURS FACTUELLES : informations dans B qui
>    contredisent A (avec citation des deux passages)
> 2. OMISSIONS SIGNIFICATIVES : éléments importants de A
>    absents de B
> 3. SURINTERPRÉTATIONS : éléments de B qui vont au-delà
>    de ce qui est dit dans A
> 4. FORMULATIONS TROMPEUSES : phrases techniquement
>    correctes mais qui donnent une impression fausse
>
> Présente les résultats dans un tableau.
>
> ===== DOCUMENT A =====
>
> [Coller le CR ici]
>
> ===== DOCUMENT B =====
>
> [Coller la mauvaise synthèse ici]
> ```

### 🔍 Écarts attendus (corrigé)

| Type | Dans la synthèse | Réalité du CR |
|------|-----------------|---------------|
| **Erreur** | « doublé le taux de conversion » | Passé de 34% à 41% (+7 points, pas un doublement) |
| **Erreur** | « notifications opérationnelles sur l'ensemble du parc » | Ne fonctionnent pas sur iOS |
| **Surinterprétation** | « moteur de règles avancé » | Simon a proposé un seuil simple, Marie a dit « MVP » |
| **Surinterprétation** | « validé à l'unanimité » | Le principe est validé mais la mécanique reste ouverte |
| **Surinterprétation** | « l'équipe technique confirme la faisabilité » | Julien dit « c'est pas un sprint, c'est un trimestre » |
| **Surinterprétation** | « recherche par mood sera intégrée » | Reporté au prochain comité pour étude |
| **Erreur** | « migration paiement terminée » | Reportée au sprint 25 faute de ressources |
| **Omission** | — | Churn restaurateurs à 8% (justification stratégique du programme) |
| **Omission** | — | Absence de Alexis sans remplaçant (risque RH majeur) |
| **Omission** | « aucun risque majeur » | Au moins 4 risques identifiables (cf. exercice 2) |

> ✅ **Point pédagogique** : Utiliser le LLM comme vérificateur est souvent plus utile que comme générateur. C'est aussi le meilleur moyen d'apprendre à lire les sorties du LLM avec un œil critique.
